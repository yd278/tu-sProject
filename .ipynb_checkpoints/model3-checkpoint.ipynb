```python
# import libs
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division  
```

```python
# set constant

mu = 1.83e-6 #dynamic viscosity of seawater
rou_p = 2.6e3 #density of particle
rou_f = 1.027e3 #density of seawater
g = 9.81 #acceleration of gravity 

#Assumptions
h = 4000 # Mean depth of Southern Ocean
current = 0.3 # velocity of ACC
diff = 0.1 * current * 1e4 # diffusivity = 0.1 * velocity * Le

# Assumptions of the distribution
stdev = np.log(2) # Standard deviation (sigma) of the grain size distribution  
mean = np.log(8e-6) # Mean the underlying normal distribution of the source grain size distribution 
```

```python
# draw 100000 diameter samples from the lognormal distribution
dia = np.random.lognormal(mean,stdev,100000)
# plot a histogram (normalized)
plt.hist(dia, 100, normed=True, align='mid') 
plt.show()
count = 0
for d in dia :
    if d > 4e-5: 
        count+=1
print(count)
```

```python
#calculate the distance

k = (rou_p-rou_f)*g/mu/18
dist = []
count = 0
for sediment in dia:

    v = k * np.square(sediment) # Stokes' law
    t = h / v # time taken 
    x = t * current # distance traveled without diffusion
    
    eta = np.random.normal(0,np.sqrt(2),1)
    x_hat = eta[0] * np.sqrt(diff*t)
    
    dist.append(x+x_hat)

plt.hist(dist, 100, normed=True, align='mid')

plt.show()
```

```python
# discard data larger then 1e5 i.e. 100 km
# and re-normalize it

TMAX = 1e5
dist_a = []
for i in dist:
    if i < TMAX:
        dist_a.append(i)

plt.hist(dist_a, 200, normed=True, align='mid')
plt.show()
```

```python
# let LE be 1e5 and do it again 
diff = 0.1 * current * 1e5

#calculate the distance

k = (rou_p-rou_f)*g/mu/18
dist = []
count = 0
for sediment in dia:

    v = k * np.square(sediment) # Stokes' law
    t = h / v # time taken 
    x = t * current # distance traveled without diffusion
    
    eta = np.random.normal(0,np.sqrt(2),1)
    x_hat = eta[0] * np.sqrt(diff*t)
    
    dist.append(x+x_hat)

plt.hist(dist, 100, normed=True, align='mid')

plt.show()
# discard data larger then 3e5 i.e. 300 km
# and re-normalize it

TMAX = 3e5
dist_a = []
for i in dist:
    if i < TMAX:
        dist_a.append(i)

plt.hist(dist_a, 200, normed=True, align='mid')
plt.show()
```
